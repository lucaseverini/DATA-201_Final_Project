```python
# 3_premier_league_analytics.ipynb

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from data201 import db_connection, df_query

# Set plot style
plt.style.use('ggplot')
sns.set(style="whitegrid")

# Connect to the database
conn = db_connection(config_file='premier_league_analytics.ini')

# Create a function to run analytical queries
def run_analytics_query(query, title=None):
    """Run an analytics query and display the results"""
    results = df_query(conn, query)
    if title:
        print(f"\n{title}:")
    display(results)
    return results

# 1. Current League Table
league_table = run_analytics_query(
    """
    SELECT * FROM vw_LeagueTable
    ORDER BY Position
    """,
    "Current League Table"
)

# 2. Home Advantage Analysis
home_advantage = run_analytics_query(
    """
    SELECT * FROM vw_HomeAdvantage
    ORDER BY HomeAdvantagePoints DESC
    LIMIT 10
    """,
    "Teams with Strongest Home Advantage"
)

# Visualize home advantage
plt.figure(figsize=(10, 6))
colors = plt.cm.viridis(np.linspace(0, 1, len(home_advantage)))
bars = plt.bar(home_advantage['TeamName'], home_advantage['HomeAdvantagePoints'], color=colors)
plt.axhline(y=0, color='r', linestyle='-', alpha=0.3)
plt.title('Home Advantage by Team (Points Difference)', fontsize=14)
plt.ylabel('Points per Game Difference (Home - Away)', fontsize=12)
plt.xlabel('Team', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 3. Goal Analysis
goal_analysis = run_analytics_query(
    """
    SELECT 
        t.TeamName,
        SUM(tms.Goals) AS TotalGoals,
        COUNT(DISTINCT tms.MatchID) AS MatchesPlayed,
        ROUND(SUM(tms.Goals) * 1.0 / COUNT(DISTINCT tms.MatchID), 2) AS GoalsPerMatch,
        SUM(CASE WHEN tms.IsHomeTeam = 1 THEN tms.Goals ELSE 0 END) AS HomeGoals,
        SUM(CASE WHEN tms.IsHomeTeam = 0 THEN tms.Goals ELSE 0 END) AS AwayGoals,
        ROUND(SUM(CASE WHEN tms.IsHomeTeam = 1 THEN tms.Goals ELSE 0 END) * 1.0 / 
              SUM(CASE WHEN tms.IsHomeTeam = 1 THEN 1 ELSE 0 END), 2) AS AvgHomeGoals,
        ROUND(SUM(CASE WHEN tms.IsHomeTeam = 0 THEN tms.Goals ELSE 0 END) * 1.0 / 
              SUM(CASE WHEN tms.IsHomeTeam = 0 THEN 1 ELSE 0 END), 2) AS AvgAwayGoals
    FROM fact_TeamMatchStats tms
    JOIN Teams t ON tms.TeamID = t.TeamID
    GROUP BY t.TeamName
    ORDER BY TotalGoals DESC
    LIMIT 10
    """,
    "Top 10 Teams by Total Goals Scored"
)

# Visualize goals per match
plt.figure(figsize=(12, 6))
x = np.arange(len(goal_analysis))
width = 0.35

fig, ax = plt.subplots(figsize=(12, 6))
home_bars = ax.bar(x - width/2, goal_analysis['AvgHomeGoals'], width, label='Home Goals per Match')
away_bars = ax.bar(x + width/2, goal_analysis['AvgAwayGoals'], width, label='Away Goals per Match')

ax.set_title('Home vs Away Goals per Match by Team', fontsize=14)
ax.set_ylabel('Goals per Match', fontsize=12)
ax.set_xticks(x)
ax.set_xticklabels(goal_analysis['TeamName'], rotation=45, ha='right')
ax.legend()
ax.grid(True, alpha=0.3)

# Add data labels
def add_labels(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.1f}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

add_labels(home_bars)
add_labels(away_bars)

plt.tight_layout()
plt.show()

# 4. Team Performance Trends
team_perf = run_analytics_query(
    """
    SELECT 
        t.TeamName,
        s.SeasonName,
        COUNT(DISTINCT tms.MatchID) AS MatchesPlayed,
        SUM(tms.Points) AS TotalPoints,
        ROUND(SUM(tms.Points) * 1.0 / COUNT(DISTINCT tms.MatchID), 2) AS PointsPerGame,
        SUM(tms.Goals) AS GoalsScored,
        SUM(tms.GoalsConceded) AS GoalsConceded,
        SUM(tms.Goals) - SUM(tms.GoalsConceded) AS GoalDifference
    FROM fact_TeamMatchStats tms
    JOIN Teams t ON tms.TeamID = t.TeamID
    JOIN Seasons s ON tms.SeasonID = s.SeasonID
    GROUP BY t.TeamName, s.SeasonName
    ORDER BY s.StartDate, PointsPerGame DESC
    """,
    "Team Performance by Season"
)

# Filter to top 6 teams for visualization
top_teams = ['Man City', 'Liverpool', 'Arsenal', 'Chelsea', 'Man United', 'Tottenham']
top_team_perf = team_perf[team_perf['TeamName'].isin(top_teams)]

# Pivot the data for visualization
pivot_ppg = pd.pivot_table(
    top_team_perf, 
    values='PointsPerGame', 
    index='SeasonName', 
    columns='TeamName',
    aggfunc='mean'
)

# Create a line chart
plt.figure(figsize=(12, 7))
for team in pivot_ppg.columns:
    plt.plot(pivot_ppg.index, pivot_ppg[team], marker='o', linewidth=2, label=team)

plt.title('Points per Game by Season - Top Teams', fontsize=14)
plt.ylabel('Points per Game', fontsize=12)
plt.xlabel('Season', fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 5. Referee Analysis
referee_analysis = run_analytics_query(
    """
    SELECT 
        r.RefereeName,
        COUNT(DISTINCT mr.MatchID) AS MatchesRefereed,
        SUM(CASE WHEN mr.Result = 'H' THEN 1 ELSE 0 END) AS HomeWins,
        SUM(CASE WHEN mr.Result = 'D' THEN 1 ELSE 0 END) AS Draws,
        SUM(CASE WHEN mr.Result = 'A' THEN 1 ELSE 0 END) AS AwayWins,
        ROUND(SUM(CASE WHEN mr.Result = 'H' THEN 1 ELSE 0 END) * 100.0 / 
              COUNT(DISTINCT mr.MatchID), 1) AS HomeWinPct,
        ROUND(SUM(CASE WHEN mr.Result = 'D' THEN 1 ELSE 0 END) * 100.0 / 
              COUNT(DISTINCT mr.MatchID), 1) AS DrawPct,
        ROUND(SUM(CASE WHEN mr.Result = 'A' THEN 1 ELSE 0 END) * 100.0 / 
              COUNT(DISTINCT mr.MatchID), 1) AS AwayWinPct
    FROM fact_MatchResult mr
    JOIN Referees r ON mr.RefereeID = r.RefereeID
    GROUP BY r.RefereeName
    HAVING MatchesRefereed >= 10
    ORDER BY HomeWinPct DESC
    LIMIT 10
    """,
    "Referees with Highest Home Win Percentage"
)

# Visualize referee home advantage
plt.figure(figsize=(14, 8))
x = np.arange(len(referee_analysis))
width = 0.25

fig, ax = plt.subplots(figsize=(14, 8))
home_bars = ax.bar(x - width, referee_analysis['HomeWinPct'], width, label='Home Win %')
draw_bars = ax.bar(x, referee_analysis['DrawPct'], width, label='Draw %')
away_bars = ax.bar(x + width, referee_analysis['AwayWinPct'], width, label='Away Win %')

ax.set_title('Match Outcomes by Referee (%)', fontsize=16)
ax.set_ylabel('Percentage of Matches', fontsize=12)
ax.set_xlabel('Referee', fontsize=12)
ax.set_xticks(x)
ax.set_xticklabels(referee_analysis['RefereeName'], rotation=45, ha='right')
ax.legend()

# Add a line for league average home win %
league_avg_query = """
SELECT 
    ROUND(SUM(CASE WHEN Result = 'H' THEN 1 ELSE 0 END) * 100.0 / COUNT(*), 1) AS AvgHomeWinPct
FROM fact_MatchResult
"""
league_avg = df_query(conn, league_avg_query).iloc[0, 0]
ax.axhline(y=league_avg, color='r', linestyle='--', alpha=0.7, label=f'League Avg Home Win ({league_avg}%)')
ax.legend()

plt.tight_layout()
plt.show()

# 6. Season Comparison
season_comparison = run_analytics_query(
    """
    SELECT 
        s.SeasonName,
        COUNT(DISTINCT fmr.MatchID) AS MatchesPlayed,
        ROUND(AVG(fmr.HomeGoals + fmr.AwayGoals), 2) AS AvgGoalsPerMatch,
        ROUND(SUM(CASE WHEN fmr.Result = 'H' THEN 1 ELSE 0 END) * 100.0 / COUNT(*), 1) AS HomeWinPct,
        ROUND(SUM(CASE WHEN fmr.Result = 'D' THEN 1 ELSE 0 END) * 100.0 / COUNT(*), 1) AS DrawPct,
        ROUND(SUM(CASE WHEN fmr.Result = 'A' THEN 1 ELSE 0 END) * 100.0 / COUNT(*), 1) AS AwayWinPct
    FROM fact_MatchResult fmr
    JOIN Seasons s ON fmr.SeasonID = s.SeasonID
    GROUP BY s.SeasonName
    ORDER BY s.StartDate DESC
    """,
    "Season-by-Season Comparison"
)

# Visualize season trends
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Goals per match
ax1.plot(season_comparison['SeasonName'], season_comparison['AvgGoalsPerMatch'], 'o-', linewidth=2)
ax1.set_title('Average Goals per Match by Season', fontsize=14)
ax1.set_ylabel('Goals per Match', fontsize=12)
ax1.set_xlabel('Season', fontsize=12)
ax1.set_xticks(range(len(season_comparison)))
ax1.set_xticklabels(season_comparison['SeasonName'], rotation=45, ha='right')
ax1.grid(True, alpha=0.3)

# Home Win %
ax2.plot(season_comparison['SeasonName'], season_comparison['HomeWinPct'], 'o-', linewidth=2)
ax2.set_title('Home Win Percentage by Season', fontsize=14)
ax2.set_ylabel('Home Win %', fontsize=12)
ax2.set_xlabel('Season', fontsize=12)
ax2.set_xticks(range(len(season_comparison)))
ax2.set_xticklabels(season_comparison['SeasonName'], rotation=45, ha='right')
ax2.set_ylim(35, 55)  # Adjust y-axis to highlight changes
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Close connection when done
conn.close()
```
